```python
import pytorch_lightning as pl
import json
from monai.utils import set_determinism
import os
from monai.data import DataLoader, Dataset
import torch
from monai.inferers import sliding_window_inference
from functools import partial
import numpy as np
import time
import matplotlib.pyplot as plt
from monai.transforms import Compose, EnsureType, AsDiscrete, Activations, LoadImaged, NormalizeIntensityd, ToTensord, AddChanneld
#os.environ["CUDA_VISIBLE_DEVICES"] = "7"
import SimpleITK as sitk
from matplotlib.colors import ListedColormap

def predict_test_inference(test_files, chosen_model = "DBAHNET", threshold  = 0.65, viz = True):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    if chosen_model == "UNET":
        from trainer_bone_unet import BONEUNET
        CKPT = "ckpt/EXP_BONE_UNET_FINAL/last.ckpt"
        model = BONEUNET().load_from_checkpoint(CKPT).to(device).eval()
    if chosen_model == "ATTENTIONUNET":
        CKPT = "ckpt/EXP_BONE_ATTENTIONUNET_2/Epoch 85-MeanDiceScore0.9611.ckpt"
        from trainer_bone_attentionunet import BONEATTENTIONUNET
        model = BONEATTENTIONUNET().load_from_checkpoint(CKPT).to(device).eval()
    if chosen_model == "UNETR":
        CKPT = "ckpt/EXP_BONE_UNETR_2/Epoch 60-MeanDiceScore0.9667.ckpt"
        from trainer_bone_unetr import BONEUNETR
        model = BONEUNETR().load_from_checkpoint(CKPT).to(device).eval()
    if chosen_model == "SWINUNETR":
        from trainer_bone_swinunetr import BONESWINUNETR
        CKPT = "ckpt/EXP_BONE_SWINUNETR_2/Epoch 60-MeanDiceScore0.9739.ckpt"
        model = BONESWINUNETR().load_from_checkpoint(CKPT).to(device).eval()
    if chosen_model =="DBAHNET":
        from trainer_bone_dbahnet import  DBAHNET
        CKPT = "ckpt/EXP_BONE_DBAHNET_96_444/Epoch 58-MeanDiceScore0.9855.ckpt"
        model = DBAHNET().load_from_checkpoint(CKPT).to(device).eval() 
    test_dataloader = get_test_dataloader(test_files)   
    roi_size = (320,320,32)
    sw_batch_size = 1
    inferer = partial(sliding_window_inference, roi_size = roi_size, sw_batch_size = sw_batch_size, predictor = model, overlap = 0.6)
    post_trans = Compose(
            [EnsureType(), Activations(sigmoid=True), AsDiscrete(threshold = threshold)]
        )
    with torch.no_grad():
         for batch_idx, batch_data in enumerate(test_dataloader):
            
            image = batch_data['image'].cuda()
            label = batch_data['label'].cuda()
            start_time = time.time()
            seg = post_trans(inferer(image))[0].detach().cpu().numpy()
            inference_time = time.time() - start_time
            print("Inference time : (s)",inference_time)
            seg_out = np.zeros((seg.shape[1], seg.shape[2], seg.shape[3]))
            seg_out[seg[0] == 1] = 0
            seg_out[seg[1] == 1] = 1
            seg_out[seg[2] == 1] = 2
            seg_out_sitk = sitk.GetImageFromArray(seg_out)
            seg_out_sitk.SetSpacing([0.005, 0.005, 0.005])
            #sitk.WriteImage(seg_out_sitk, os.path.join(OUT_DIR, f"pred_{count}.nii.gz"))
            #count +=1
            cmap_seg = ListedColormap(['black','darkorange', 'dodgerblue'], name='custom_segmentation')

            if viz :
                image = image[0][0].detach().cpu().numpy()
                label = label[0].detach().cpu().numpy()
                print(label.shape)

                slice_num = 5
                plt.figure("image", (18, 6))
                plt.subplot(1, 3, 1)
                plt.title("slice")
                plt.imshow(image[:, :, slice_num], cmap="gray")
                plt.subplot(1, 3, 2)
                plt.title("segmentation")
                plt.imshow(seg_out[:, :, slice_num], cmap =cmap_seg)
                plt.subplot(1, 3, 3)
                plt.title("label")
                plt.imshow(label[:, :, slice_num], cmap =cmap_seg)
                plt.show()
```

```python
import pytorch_lightning as pl
import json
from monai.utils import set_determinism
import os
from monai.data import DataLoader, Dataset
import torch
from monai.inferers import sliding_window_inference
from functools import partial
import numpy as np
import time
import matplotlib.pyplot as plt
#os.environ["CUDA_VISIBLE_DEVICES"] = "7"
import SimpleITK as sitk
from matplotlib.colors import ListedColormap
from monai.transforms import (
    Transform,
    Compose,
    LoadImaged,
    RandFlipd,
    NormalizeIntensityd,
    RandScaleIntensityd,
    RandCropByPosNegLabeld,
    ToTensord,
    AddChanneld,
    MapTransform,
    Affined,
    Rand3DElasticd,
    RandAdjustContrastd,
    SpatialPadd,
    # RandShiftIntensityd,
    # Orientationd,
    # ScaleIntensityRanged,
    # SpatialPadd,
    # CropForegroundd,
    # RandSpatialCropSamplesd,
    

)   
class ConvertToMultiChannelForBoneClassesd(MapTransform):
    """
    Convert labels to multi channels based on bone classes:
    The possible classes are BG 0, Cortical 1, and Trabecular 2.
    """

    def __call__(self, data):
        d = dict(data)
        for key in self.keys:
            result = []
            result.append(d[key] == 0)
            
            # Cortical class
            result.append(d[key] == 1)
            
            # Trabecular class
            result.append(d[key] == 2)
            # Stack the results
            multi_channel_label = np.stack(result, axis=0).astype(np.float32)

            d[key] = multi_channel_label
            
        return d
        
        
def get_test_dataloader(test_files):
    test_transform = Compose(
        [
            LoadImaged(keys=["image", "label"]),
            ConvertToMultiChannelForBoneClassesd(keys = ['label']),
            AddChanneld(keys = ["image"]),
            NormalizeIntensityd(keys = "image",
                               nonzero = True,
                               channel_wise = True),
            ToTensord(keys=["image", "label"]),
        ]
    )
    test_ds = Dataset(data=test_files, transform=test_transform)
    test_loader = DataLoader(test_ds, batch_size=1, shuffle=False)

    return test_loader
def evaluate_test(test_files, chosen_model = "DBAHNET", threshold  = 0.65, viz = True):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    if chosen_model == "UNET":
        from trainer_bone_unet import BONEUNET
        CKPT = "ckpt/EXP_BONE_UNET_FINAL/last.ckpt"
        model = BONEUNET().load_from_checkpoint(CKPT).to(device).eval()
    if chosen_model == "ATTENTIONUNET":
        CKPT = "ckpt/EXP_BONE_ATTENTIONUNET_2/Epoch 85-MeanDiceScore0.9611.ckpt"
        from trainer_bone_attentionunet import BONEATTENTIONUNET
        model = BONEATTENTIONUNET().load_from_checkpoint(CKPT).to(device).eval()
    if chosen_model == "UNETR":
        CKPT = "ckpt/EXP_BONE_UNETR_2/Epoch 60-MeanDiceScore0.9667.ckpt"
        from trainer_bone_unetr import BONEUNETR
        model = BONEUNETR().load_from_checkpoint(CKPT).to(device).eval()
    if chosen_model == "SWINUNETR":
        from trainer_bone_swinunetr import BONESWINUNETR
        CKPT = "ckpt/EXP_BONE_SWINUNETR_2/Epoch 60-MeanDiceScore0.9739.ckpt"
        model = BONESWINUNETR().load_from_checkpoint(CKPT).to(device).eval()
    if chosen_model =="DBAHNET":
        from trainer_bone_dbahnet import  DBAHNET
        CKPT = "ckpt/EXP_BONE_DBAHNET_96_444/Epoch 58-MeanDiceScore0.9855.ckpt"
        model = DBAHNET().load_from_checkpoint(CKPT).to(device).eval() 
    test_dataloader = get_test_dataloader(test_files)   

    trainer = pl.Trainer(accelerator= "gpu", precision=16)

    #print("Validation set")
    #trainer.test(model, dataloaders = val_dataloader)
    print("Testing set")
    trainer.test(model, dataloaders = test_dataloader)
```

```python
evaluate_test(test_files, chosen_model = "DBAHNET", threshold  = 0.65, viz = True)
```
